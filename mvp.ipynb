```python
# Import Data

def preprocessFile( filePath, startingDate ):
    # Returns the preprocessed data (includes z-score)
    data = pd.read_csv(filePath)
    # filter out data out of our date range
    data = data[data['Date'] >= startingDate]
    data['Date'] = pd.Series([pd.to_datetime(date) for date in data['Date']])
    
    # Compute z-score

    mean = np.mean(data['Value'])
    std = np.std(data['Value'])
    zscore = (data['Value']-mean)/std
    
    data['zscore'] = zscore
    
    return data
```

```python
d1 = preprocessFile('data/USGOVT.csv', '2004-01-01')
d2 = preprocessFile('data/ASPNHSUS.csv', '2004-01-01')
```

```python
# core pandas correlate function
# output (default) is the pearson coefficient

print d1['zscore'].corr(d2['zscore'])
print d1['zscore'].cov(d2['zscore'])
```

```python
# plot the two data sets
# Uses z-score to better visualize the correlation

plt.plot(d1['Date'], d1['zscore'])
plt.plot(d2['Date'], d2['zscore'])
```

```python
# plot the data correlation (data set 1 in x, data set 2 in y)

plt.scatter(d1['zscore'],d2['zscore'])

# adds a trend line
# if the data sets where perfectly correlated, the line would fit nicely

plt.plot(d1['zscore'],d1['zscore'])
```
